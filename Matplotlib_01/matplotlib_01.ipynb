```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,15,10,5,10]
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.show()
```

```python
import numpy as np
x = np.linspace(0,10,100) #100 number, 0 START 10 END
y = np.cos(x)#new data 
z = np.sin(x)
a = np.tan(x)
plt.plot (x,y,z,a)
plt.title("Line plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.legend(['data','sin','tan'])
plt.show()
```

```python
x = np.random.rand(100) #start 0 end 1
y = np.random.rand(100)

plt.scatter(x,y,color='red',s=20,alpha=0.5)
plt.title("Scatter Plot")
plt.xlabel("date")
plt.ylabel("sales")
plt.grid(True,alpha=0.3)
plt.show()
```

```python
#bar plot
categories = ['apple','banana','orange','grapes']
sales =[25,40,32,20]

plt.bar(categories,sales,color =['red','yellow','orange','green'])
plt.grid(True,alpha=0.3)
plt.title("Sales of Fruits")
plt.xlabel("categories")
plt.ylabel("sales")
plt.show()
```

```python
#bar plot
categories = ['apple','banana','orange','grapes']
# sales =[25,40,32,20]
sales_23 = [50,40,20,90]
sales_24 = [45,23,19,80]
sales_26 = [10,15,20,23]

x = np.arange(len(categories))

plt.bar(x-0.1,sales_23,label="sales 2023",width=0.4)
plt.bar(x-0.2,sales_24,label="sales 2024",width=0.4)
plt.bar(x-0.3,sales_26,label="sales 2026",width=0.4)
plt.grid(True,alpha=0.3)
plt.title("Sales of Fruits")
plt.xticks(x,categories)
plt.xlabel("sales")
plt.ylabel("categories")
plt.legend()
plt.show()
```
